```python
import os
from bs4 import BeautifulSoup
from selenium import webdriver
from selenium.webdriver.common.by import By
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from time import sleep
from datetime import datetime
import argparse
import csv
from selenium.webdriver.common.keys import Keys
```

```python
def get_historic_cotization(url, start, end, t):
    var1 = start
    var2 = end
    topic = t
    #start Webdriver using Chrome driver
    driver = webdriver.Chrome(r"C:\Users\Sierra\Downloads\chromedriver_win32\chromedriver.exe")
    driver.get(url)
    WebDriverWait(driver, 8)
    
    #Find calender date picker using css selector
    dates = driver.find_element_by_css_selector('#flatDatePickerCanvasHol > div:nth-child(2)')
    dates.click()

    # dynamically generate displayed variables/testing
    _vars = driver.find_element_by_id('curr_table').find_elements_by_tag_name('th')
    variables = [var.text.lower() for var in _vars]
    print(variables)

    #inputing dates into fields
    startDate = driver.find_element_by_id('startDate')
    startDate.clear()
    startDate.send_keys(var1)
    
    endDate = driver.find_element_by_id('endDate')
    endDate.clear()
    endDate.send_keys(var2)
    
    #Apply changes by finding tag and class name
    apply_changes = driver.find_element_by_class_name('ui-datepicker-buttonpane').find_element_by_tag_name('a')
    apply_changes.click()
    
    #Sort date by fing css selector
    date = driver.find_element_by_css_selector('#curr_table > thead > tr > th.first.left.noWrap.pointer > span')
    date.click()

    #Use bs4 to scrape html page
    soup = BeautifulSoup(driver.page_source, "lxml")
    table = soup.find(class_='genTbl closedTbl historicalTbl')
    
    #find all rows & col in order to extrat information
    rows = table.find_all('td')
    counter=0
    
    #Open csv to write information using dictionary
    d = {'Date':'','Price':'','Open':'','High':'','Low':'', 'Vol':'', 'Change':''}
    f = csv.writer(open("%s\\%s-%s-%s.csv" % (os.getcwd(), var1.replace('/','.'), var2.replace('/','.'), topic), 'w', newline=''))
    f.writerow(['Date','Price','Open', 'High', 'Low', 'Vol.', 'Change %'])
    
    #Write rows to csv file accordingly
    for tag in rows:
        tag_string = str(tag.string)
        if counter == 0:
            d['Date'] = tag_string
            counter = counter+1
            continue
        elif counter == 1:
            d['Price'] = tag_string
            counter = counter+1
            continue
        elif counter == 2:
            d['Open'] = tag_string
            counter = counter+1
            continue
        elif counter == 3:
            d['High'] = tag_string
            counter = counter+1
            continue
        elif counter == 4:
            d['Low'] = tag_string
            counter = counter+1
            continue
        elif counter == 5:
            d['Vol'] = tag_string
            counter = counter+1
            continue
        elif counter == 6:
            d['Change'] = tag_string
            counter = counter+1
            continue
        else:
            counter=1
            f.writerow([d['Date'], d['Price'], d['Open'], d['High'], d['Low'], d['Vol'], d['Change']])
            d['Date'] = tag_string
    
    
```
