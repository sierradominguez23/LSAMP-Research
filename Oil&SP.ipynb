```python
import pandas as pd
```

```python
oil1 = pd.read_csv('01.01.2015-07.01.2015-Oil1.csv')
oil2 = pd.read_csv('07.01.2015-01.01.2016-Oil2.csv')
oil3 = pd.read_csv('01.01.2016-07.01.2016-Oil3.csv')
oil4 = pd.read_csv('07.01.2016-01.01.2017-Oil4.csv')
oil5 = pd.read_csv('01.01.2017-04.02.2017-Oil5.csv')
```

```python
sp1 = pd.read_csv('01.01.2015-07.01.2015-SP1.csv')
sp2 = pd.read_csv('07.01.2015-01.01.2016-SP2.csv')
sp3 = pd.read_csv('01.01.2016-07.01.2016-SP3.csv')
sp4 = pd.read_csv('07.01.2016-01.01.2017-SP4.csv')
sp5 = pd.read_csv('01.01.2017-04.02.2017-SP5.csv')
```

```python
oil = pd.concat([oil1, oil2, oil3, oil4, oil5])
```

```python
oil.reset_index()
```

```python
oil.drop('Open',axis=1,inplace=True)
oil.drop('High',axis=1,inplace=True)
oil.drop('Low',axis=1,inplace=True)
oil.drop('Vol.',axis=1,inplace=True)
oil.drop('Change %',axis=1,inplace=True)
```

```python
oil.drop('Change %', axis=1, inplace=True)
```

```python
oil.head()
```

```python
oil['Oil Price'] = oil['Price']
```

```python
oil.head()
```

```python
oil.drop('Price', axis=1, inplace=True)
```

```python
oil.head()
```

```python
sp = pd.concat([sp1, sp2, sp3, sp4, sp5])
```

```python
sp.reset_index()
```

```python
sp.drop('Open',axis=1,inplace=True)
sp.drop('High',axis=1,inplace=True)
sp.drop('Low',axis=1,inplace=True)
sp.drop('Vol.',axis=1,inplace=True)
sp.drop('Change %',axis=1,inplace=True)
```

```python
sp.head()
```

```python
sp['S&P Price'] = sp['Price']
```

```python
sp.head()
```

```python
sp.drop('Price', axis=1, inplace=True)
```

```python
sp.head()
```

```python
df = pd.merge(oil, sp, how='inner', on='Date')
```
